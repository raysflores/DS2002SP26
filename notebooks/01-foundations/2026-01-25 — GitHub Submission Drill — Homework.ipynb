
# DS2002 — GitHub Lab 01  
## Your First Repository, Files, and Commits
**Course:** DS2002 — Data Science Systems  
**Purpose:** Learn GitHub by *doing*, not watching

---

This lab assumes:
- You have **never** used GitHub seriously before
- You completed the GitHub Fundamentals lecture
- You are willing to go step by step


By the end of this lab, you will:
- create a **private GitHub repository**
- add and edit multiple files
- make multiple commits
- invite your instructor
- submit a GitHub URL in Canvas

This repository will become the foundation for your work in this course.



# What You Are Building

You are creating a **course workspace repository**.

Think of this repository as:
- your DS2002 folder
- with memory
- and checkpoints

Everything you do in this lab mirrors how professionals use GitHub in real projects.



# Step 1 — Create a Private Repository

1. Go to https://github.com and log in
2. Click the **+** icon (top right)
3. Choose **New repository**

Fill in the form carefully:

- Repository name:  
  `DS2002-YourLastName`
- Description:  
  `Private DS2002 coursework repository`
- Visibility:  
  **Private**
- Initialize repository:
  ✔ Add a README file

Click **Create repository**



# Step 2 — Invite Your Instructor

Because this repository is private, your instructor cannot see it unless invited.

1. In your repository, click **Settings**
2. Click **Collaborators**
3. Click **Add people**
4. Enter your instructor’s GitHub username (provided in class)
5. Send the invitation

This step is required for grading.



# Step 3 — Understand the Repository View

Take a moment to look around.

You should see:
- a file called `README.md`
- a list of commits (History)
- buttons for adding files

This is your project home.



# Step 4 — Edit Your README (First Commit)

1. Click `README.md`
2. Click the **pencil (Edit)** icon

Replace the contents with the following (customize your name):

# DS2002 Coursework

Name: Your Name  
Course: DS2002 — Data Science Systems  

This repository contains my labs, notebooks, and assignments for DS2002.

Scroll down.

Commit message:
`Initialize DS2002 repository with README`

Click **Commit changes**



# Step 5 — Create a Folder Structure

Professional repositories are organized.

We will create a basic structure that you will reuse all semester.

1. Click **Add file → Create new file**
2. In the filename box, type:

`labs/README.md`

GitHub will automatically create the folder.

In the file contents, write:

# Labs

This folder contains lab work for DS2002.

Commit message:
`Add labs folder`

Click **Commit changes**



# Step 6 — Add a Notes File

Now we will add a plain text file.

1. Click **Add file → Create new file**
2. Filename:

`notes/week01.txt`

In the file, write a short paragraph answering:

- What is GitHub?
- Why are we using it in this course?

There is no “right” answer. Use your own words.

Commit message:
`Add week 1 notes`

Commit changes.



# Step 7 — Edit an Existing File (Second Commit)

Open `notes/week01.txt` again.

Add one more sentence at the bottom:
- What confused you most about GitHub so far?

Commit message:
`Update week 1 notes with reflection`

Commit changes.



# Step 8 — View Your Commit History

Click the **Commits** link near the top of the repository.

You should see:
- multiple commits
- each with a message
- each with a timestamp

This history is the most important feature of GitHub.



# Step 9 — Verify Instructor Access

Make sure:
- the repository is **private**
- your instructor has been invited
- the invitation shows as accepted (or pending)

If your instructor cannot see your repo, it cannot be graded.



# Step 10 — Submit the Repository URL

Copy the URL of your repository.

It will look like:
https://github.com/your-username/DS2002-YourLastName

Submit this URL in the Canvas assignment.

Do **not** upload files to Canvas.



# Submission Checklist

Before submitting, confirm:

✔ Repository is private  
✔ Instructor invited as collaborator  
✔ README.md edited  
✔ labs/ folder exists  
✔ notes/week01.txt exists  
✔ Multiple commits visible  

If all are true, you are done.



# Why This Lab Matters

This lab is not about GitHub features.

It is about learning to:
- work incrementally
- leave a clean history
- organize your work
- share progress safely

These habits will matter far beyond this course.



# You Are Finished

You now have:
- a real GitHub repository
- real commits
- real structure

Every future lab builds on this.

Nice work.
